```python
def real_date(x):
    """ функция на входе принимает строчную переменную,
    проверяет корректность года и в случае, 
       - если год корректный- превращает дату в тип Timestamp
       - если год некорректный ( меньше 1920 или больше 2020):
         выдает token формата Timestamp '1800-01-01'
    """     
    xx=x.split('-')
    if int(xx[0])<1920  or int(xx[0])>2020:
        return pd.Timestamp('1800-01-01')
    else:
        return pd.Timestamp(x)
stud_drop_na['birthday_date']=stud_drop_na['birthday'].apply(real_date)
print(mark1.format('вот такие интересные года рождения есть у студентов. Некоторые староваты, некоторые из будущего'))
print(mark1.format('но есть подозрение, что неслучайные это ошибки.... Почему битые данные только по году?'))
students_clear=stud_drop_na[stud_drop_na['birthday_date']>pd.Timestamp('1800-01-01')]
stud_drop_na[stud_drop_na['birthday_date']==pd.Timestamp('1800-01-01')][['id_','id','birthday']]
```
